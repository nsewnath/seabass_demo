```python
from SB_support import *
import pandas as pd
```

```python
data = readSB(filename = "./original_data/chl-AL9807.sb", no_warn = True)
```

```python
data.data
```

```python
df = pd.DataFrame(data.data, columns=data.data.keys())
```

```python
df
```

```python
#df1['key'] = df1['Order_ID'] + '_' + df1['Date'].apply(str)  # .apply(str) is new

#lambda function that puts a 0 in front of numbers that are less than 10 


df = df.assign(mod_minute = df["minute"].apply(str))
df['mod_minute'] = df['mod_minute'].apply(lambda x: x.zfill(2))

df = df.assign(complete_time = df["hour"].apply(str) + ":" + df["mod_minute"])
df = df.assign(complete_date = df["day"].apply(str) + "/" + df["month"].apply(str) + "/" + df["year"].apply(str))
```
